```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
plt.style.use('ggplot')
```

```python
AAPL = pd.read_csv ("NSE_BANKING_SECTOR.csv")#("GME_stock.csv")
```

```python
print(AAPL.isna().sum())
AAPL
```

```python
#VISUALIZATION
plt.figure(figsize = (13,5))
plt.plot(AAPL['CLOSE'], label = "adj CLOSE PRICE")
plt.xlabel('FROM 2016 01 01 - 2021 05 28')
plt.ylabel("CLOSE PRICE USD")
plt.legend(loc = 'upper left')
plt.show()
```

```python
SMA30 = pd.DataFrame()
SMA30['Cclose'] = AAPL["CLOSE"].rolling(window = 30).mean()
SMA100 = pd.DataFrame()
SMA100['Cclose'] = AAPL["CLOSE"].rolling(window = 100).mean()
```

```python
plt.figure(figsize = (7,7))
plt.plot(AAPL['CLOSE'], label = "adj CLOSE PRICE")
plt.plot(SMA30["Cclose"], label = "SMA30")
plt.plot(SMA100["Cclose"], label = "SMA100")
plt.xlabel('FROM 2016 01 01 - 2021 05 28')
plt.ylabel("CLOSE PRICE USD")
plt.legend(loc = 'upper left')
plt.show()
# if sma30 are crosing sma100 from top thats mean we should short
# if  sma30 are crosing sma100 from bottom we should go long
```

```python
df = pd.DataFrame()
df['cclose'] = AAPL['CLOSE']
df['SMA30'] = SMA30['Cclose']
df['SMA100'] = SMA100['Cclose']
df['sell_signal'] = 0
df['buy_signal'] = 0

print(df.isnull().sum())
df = df.fillna(0)
df
```

```python
def dual_sma(df):
    buy_signal_price = []
    sell_signal_price = []
    flag = 0
    
    for i in range(len(df)):
        if df["SMA30"][i] > df["SMA100"][i]:
            if flag != 1:
                buy_signal_price.append(df["cclose"][i])
                sell_signal_price.append(np.nan)
                flag = 1
            else:
                buy_signal_price.append(np.nan)
                sell_signal_price.append(np.nan)
        elif df["SMA30"][i] < df["SMA100"][i]:
            if flag != 1:
                buy_signa_price.append(np.nan)
                sell_signal_price.append(df["cclose"][i])
                flag = -1
            else:
                buy_signal_price.append(np.nan)
                sell_signal_price.append(np.nan)
            
        else:
            buy_signal_price.append(np.nan)
            sell_signal_price.append(np.nan)
    return(buy_signal_price, sell_signal_price)
        
```

```python
kek = dual_sma(df)
df['buy_signal'] = kek[0]
df['sell_signal'] = kek[1]
```

```python
# visualiization

plt.figure(figsize = (7,7))
plt.plot(AAPL['CLOSE'], label = "CLOSE PRICE", alpha = 0.6)
plt.plot(SMA30["Cclose"], label = "SMA30", linewidth = 3)
plt.plot(SMA100["Cclose"], label = "SMA100",linewidth = 3)
plt.scatter(df.index, df['buy_signal'], label = 'buy', color = 'green',
            marker = '^', linewidth = 5)
plt.scatter(df.index, df['sell_signal'], label = 'sell', color = 'red',
            marker = '<', linewidth = 5)
plt.xlabel('FROM 2016 01 01 - 2021 05 28')
plt.ylabel("CLOSE PRICE USD")
plt.legend(loc = 'upper left')
plt.show()
```
